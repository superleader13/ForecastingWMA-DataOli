```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df=pd.read_excel("data1th.xlsx", header=2)   #import data excel 
df
# df
```

```python
# drop digunakan untuk memotong data kolom
df.drop('No', axis=1, inplace=True)

# df.rename(columns = {'No':'weights'}, inplace = True)
df.rename(columns = {'Data aktual':'barangkeluar'}, inplace = True)
df
```

```python
#hitung WMA Forecast
# from numpy_ext import rolling_apply as rolling_apply_ext
weights = np.array([1,2,3])
df['WMA']=(df['barangkeluar']
            .rolling(3)
            .apply(lambda barangkeluar: np.dot(barangkeluar,weights)/weights.sum(), raw=True))


# wma=df.barangkeluar.rolling(window=4).apply(wma, raw=False)
# df['WMA']=df.rolling(window=4).apply(lambda WMA:np.random.choice(WMA, size=1))
df
```

```python
#hitung absolut Error(MAE)
n = len(df['barangkeluar'])

for i in range (n):
    df['Error'] = (df['barangkeluar']-df['WMA'])
    df['MAE'] = np.abs(df['Error'])    #bernilai/mendekati 0 akurat
    
    df['MAPE']= (df['Error']/n*100)    #<10 sampek 50 baik
df
```
